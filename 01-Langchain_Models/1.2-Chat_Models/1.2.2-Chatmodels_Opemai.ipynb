### Temparature in Language Models

Temperature controls the randomness of a model’s output.  
- **Lower values (0.0 – 0.3):** More deterministic, predictable responses  
- **Higher values (0.7 – 1.5):** More random, creative, and diverse responses  

#### Recommended Temperature Ranges

| Use Case                                       | Recommended Temperature |
|------------------------------------------------|--------------------------|
| Factual answers (math, code, facts)            | 0.0 – 0.3                |
| Balanced response (general QA, explanations)   | 0.5 – 0.7                |
| Creative writing, storytelling, jokes          | 0.9 – 1.2                |
| Maximum randomness (wild ideas, brainstorming) | 1.5+                     |


```python
## Instantiate chatmodel
model = ChatOpenAI(
    model="gpt-4.1-nano",
    temperature=0
)

## Call the chatbot model
response = model.invoke("Write a poem about bird")

print(response.content)
```

```python
## Instantiate chatmodel
model = ChatOpenAI(
    model="gpt-4.1-nano",
    temperature=1.1         # creative mode
)

## Call the chatbot model
response = model.invoke("Write a poem about bird")

print(response.content)
```

### max_completion_tokens
It restricts the output within a limited range. Very useful for paid models.

```python
## Instantiate chatmodel
model = ChatOpenAI(
    model="gpt-4.1-nano",
    temperature=1.1,         # creative mode
    max_completion_tokens=20
)

## Call the chatbot model
response = model.invoke("Write a poem about bird")

print(response.content)
```
